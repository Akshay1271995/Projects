```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns',100)
```

```python
movies_df = pd.read_csv('movies.csv',usecols=['movieId','title'])
rating_df=pd.read_csv('ratings.csv',usecols=['userId', 'movieId', 'rating'])
```

```python
rating_df.isnull().sum()
```

```python
df = pd.merge(rating_df,movies_df,on='movieId')
df.head()
```

```python
df.dtypes
```

```python
combine_movie_rating = df.dropna(axis = 0, subset = ['title'])
```

```python
Movie_Rating_Count = combine_movie_rating.groupby(by = ['title'])['rating'].count().reset_index().rename(columns = {'rating': 'TotalRatingsCounts'})
```

```python
Movie_Rating_Count
```

```python
rating_with_totalRatingCount = combine_movie_rating.merge(Movie_Rating_Count, left_on = 'title', right_on = 'title', how = 'left')
rating_with_totalRatingCount.head()
```

```python
print(Movie_Rating_Count['TotalRatingsCounts'].describe())
```

```python
rating_popular_movie=rating_with_totalRatingCount[(rating_with_totalRatingCount.TotalRatingsCounts>=50)]
```

```python
rating_popular_movie
```

```python
## First lets create a Pivot matrix

movie_features_df=rating_popular_movie.pivot_table(index='title',columns='userId',values='rating').fillna(0)
movie_features_df
```

```python
from scipy.sparse import csr_matrix

movie_features_df_matrix = csr_matrix(movie_features_df.values)

```

```python
#p=2 'euclidian Distance', unsupervised learning
from sklearn.neighbors import NearestNeighbors


model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute')
model_knn.fit(movie_features_df_matrix)
```

```python
movie_features_df.shape
```

```python
movie_features_df.head()
```

```python
query_index = np.random.choice(movie_features_df.shape[0])
print(query_index)
distances, indices = model_knn.kneighbors(movie_features_df.iloc[query_index,:].values.reshape(1, -1), n_neighbors = 10)
```

```python
distances
```

```python
indices
```

```python
for i in range(0, len(distances.flatten())):
    if i == 0:
        print('Recommendations for {0}:\n'.format(movie_features_df.index[query_index]))
    else:
        print('{0}: {1}, with distance of {2}:'.format(i, movie_features_df.index[indices.flatten()[i]], distances.flatten()[i]))
```
